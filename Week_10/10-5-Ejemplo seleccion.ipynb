```python
from sklearn.datasets import load_boston
import pandas as pd

# Cargar el conjunto de datos de Boston
boston_dataset = load_boston()
data = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
data['MEDV'] = boston_dataset.target  # Añadimos la variable objetivo
```

Significado de las variables:

 - CRIM     per capita crime rate by town
 - ZN       proportion of residential land zoned for lots over 25,000 sq.ft.
 - INDUS    proportion of non-retail business acres per town
 - CHAS     Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)
 - NOX      nitric oxides concentration (parts per 10 million)
 - RM       average number of rooms per dwelling
 - AGE      proportion of owner-occupied units built prior to 1940
 - DIS      weighted distances to five Boston employment centres
 - RAD      index of accessibility to radial highways
 - TAX      full-value property-tax rate per $10,000
 - PTRATIO  pupil-teacher ratio by town
 - B        1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
 - LSTAT    % lower status of the population
 - MEDV     Median value of owner-occupied homes in $1000's

```python
# Dividir los datos en variables independientes (X) y la variable dependiente (y)
X = data.drop('MEDV', axis=1)
y = data['MEDV']

# Dividir los datos en conjuntos de entrenamiento y prueba
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler

# Estandarizar las características
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)  # Nota: Solo transformamos el conjunto de prueba.

X_train_scaled
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Ajustar el modelo de regresión lineal
model = LinearRegression()
model.fit(X_train_scaled, y_train)

# Predecir y calcular la suma de cuadrados residuales
y_pred = model.predict(X_test_scaled)
rss = np.sum((y_test - y_pred) ** 2)


# Calcular AIC para la regresión lineal
n = X_test_scaled.shape[0]  # número de observaciones
k = X_test_scaled.shape[1]  # número de parámetros
aic = n * np.log(rss / n) + 2 * k

# Calcular BIC para la regresión lineal
bic = n * np.log(rss / n) + np.log(n) * k

print("AIC: ", aic)
print("BIC: ", bic)

mse = mean_squared_error(y_test, y_pred)
r2_rfe = r2_score(y_test, y_pred)
print("MSE: ", mse)
print("R2: ", r2_rfe)
```

```python
# Seleccionar las variables con mayor correlación con la variable objetivo
correlations = data.corr().abs()['MEDV'].sort_values(ascending=False)
print(correlations)
```

```python

# Considerar variables con una correlación por encima de un umbral, por ejemplo, 0.5
features = correlations[correlations > 0.5].index.tolist()
features.remove('MEDV')  # Eliminar la variable objetivo
print(features)
```

```python

X_train_corr = X_train[features]
X_test_corr = X_test[features]

# Ajustar el modelo de regresión lineal
model = LinearRegression()
model.fit(X_train_corr, y_train)

# Predecir y calcular la suma de cuadrados residuales
y_pred = model.predict(X_test_corr)
rss = np.sum((y_test - y_pred) ** 2)

print("print SRR: ", rss)
# Calcular AIC para la regresión lineal
n = X_test_corr.shape[0]  # número de observaciones
k = X_test_corr.shape[1]  # número de parámetros
aic = n * np.log(rss / n) + 2 * k

# Calcular BIC para la regresión lineal
bic = n * np.log(rss / n) + np.log(n) * k

print("AIC: ", aic)
print("BIC: ", bic)

mse = mean_squared_error(y_test, y_pred)
r2_rfe = r2_score(y_test, y_pred)
print("MSE: ", mse)
print("R2: ", r2_rfe)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE

# Instanciar el modelo de regresión lineal
model = LinearRegression()

# Utilizar RFE (Recursive Feature Elimination) para seleccionar variables
selector = RFE(model, n_features_to_select=5)
selector = selector.fit(X_train, y_train)

# Seleccionar las variables
features = X_train.columns[selector.support_].tolist()
print(features)
```

```python

# Mantener solo las variables seleccionadas
X_train_rfe = selector.transform(X_train)
X_test_rfe = selector.transform(X_test)
```

```python
# Ajustar el modelo de regresión lineal con las características seleccionadas por RFE
model_rfe = LinearRegression()
model_rfe.fit(X_train_rfe, y_train)



y_pred_rfe = model_rfe.predict(X_test_rfe)
mse_rfe = mean_squared_error(y_test, y_pred_rfe)
r2_rfe = r2_score(y_test, y_pred_rfe)

print("MSE con RFE: ", mse_rfe)
print("R^2 con RFE: ", r2_rfe)
```

```python
from sklearn.feature_selection import SelectKBest, f_regression

# Seleccionar las k mejores características según el test ANOVA
k = 5
selector_kbest = SelectKBest(f_regression, k=k)
X_train_kbest = selector_kbest.fit_transform(X_train_scaled, y_train)
X_test_kbest = selector_kbest.transform(X_test_scaled)

# Podemos ver qué características han sido seleccionadas
features_kbest = selector_kbest.get_support(indices=True)
print("Características seleccionadas por SelectKBest: ", boston_dataset.feature_names[features_kbest])
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS

# Definir Sequential Forward Selection (sfs)
sfs = SFS(LinearRegression(),
          k_features=5,
          forward=True,
          floating=False,
          scoring='neg_mean_squared_error',
          cv=0)  # Establecemos cv=0 para no usar validación cruzada y acelerar el proceso

# Ajustar SFS al conjunto de entrenamiento escalado
sfs = sfs.fit(X_train_scaled, y_train)

# Seleccionar las características indicadas por SFS
X_train_sfs = sfs.transform(X_train_scaled)
X_test_sfs = sfs.transform(X_test_scaled)

# Podemos ver qué características han sido seleccionadas
features_sfs = list(sfs.k_feature_idx_)
print("Características seleccionadas por SFS: ", boston_dataset.feature_names[features_sfs])
```
