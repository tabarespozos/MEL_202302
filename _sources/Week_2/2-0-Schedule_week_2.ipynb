## Semana 2. Probabilidad y distribuciones Probabilidad y Distribuciones



**Objetivo**
Al final de esta semana, los estudiantes deben ser capaces de entender los conceptos básicos de probabilidad, tipos de distribuciones y sus aplicaciones.

**Programa:**

*Introducción a la Probabilidad*
- Definición de Probabilidad
- Teoría Básica de la Probabilidad: Sucesos, Espacio Muestral
- Reglas de Probabilidad: Suma, Multiplicación, Probabilidad Condicional, Teorema de Bayes

*Distribuciones Discretas de Probabilidad*.
- Comprensión de las distribuciones discretas
- Distribución Binomial
- Distribución de Poisson

*Distribuciones de probabilidad continuas*
- Entender las distribuciones continuas
- Distribución Normal
- Distribución Exponencial

*Aplicaciones de las Distribuciones de Probabilidad*
- Comprensión de la importancia de las distribuciones
- Aplicaciones reales de varias distribuciones
