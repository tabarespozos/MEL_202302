### 1. Introducción

La correlación es un concepto fundamental en estadística y ciencia de datos. Cuantifica el grado en que dos variables están relacionadas entre sí.

Comprender la correlación es crucial para explorar datos, formular hipótesis y construir modelos.

Existen varios tipos de coeficientes de correlación, pero los más utilizados son el coeficiente de correlación de Pearson, el coeficiente de correlación de rangos de Spearman y el coeficiente de correlación de rangos de Kendall.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

X1 = [1,2,3,4,5]
y1 = [1,2,3,4,5]

X2 = [1,2,3,4,5]
y2 = [1,1,1,1,1]

X3 = [5,4,3,2,1]
y3 = [1,2,3,4,5]

```

### 2. Coeficiente de correlación de Pearson

El coeficiente de correlación de Pearson (r) mide la relación lineal entre dos variables. Oscila entre -1 (correlación lineal negativa perfecta) y 1 (correlación lineal positiva perfecta). Un coeficiente de correlación de 0 significa que no existe relación lineal entre las variables.

La fórmula del coeficiente de correlación de Pearson es:
$$
r = \frac{{N\sum xy - (\sum x)(\sum y)}}{{\sqrt{{[N\sum x^2 - (\sum x)^2][N\sum y^2 - (\sum y)^2]}}}}
$$

where:
- $N$ es el número de pares de puntuaciones
- $\sum xy$ es la suma de los productos de las puntuaciones emparejadas
- $\sum x$ y $\sum y$ son las sumas de las puntuaciones x e y,
- $\sum x^2$ y $\sum y^2$ son las sumas de las puntuaciones x e y al cuadrado.




Código Python para calcular la r de Pearson:



```python
import numpy as np

# create two variables
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([2, 3, 4, 5, 6, 7])

# calculate parts of the formula
N = len(x)
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x_square = np.sum(x*x)
sum_y_square = np.sum(y*y)

# calculate Pearson Correlation Coefficient
r = (N * sum_xy - sum_x * sum_y) / np.sqrt((N * sum_x_square - sum_x**2) * (N * sum_y_square - sum_y**2))

print("Pearson correlation coefficient is", r)
```

### 3. Coeficiente de correlación de rango de Spearman

El coeficiente de correlación por rangos de Spearman (ρ, también denominado $r_s$) es una medida no paramétrica de la correlación, lo que significa que no asume una relación lineal ni distribuciones normales. En su lugar, evalúa lo bien que una función monótona arbitraria puede describir la relación entre dos variables, sin hacer ninguna otra suposición sobre la naturaleza particular de la relación entre las variables.

La fórmula para calcular el coeficiente de correlación de rango de Spearman es:
$$
\rho = 1 - \frac{{6\sum d^2}}{{n(n^2 - 1)}}
$$


donde:
- $d_i$ es la diferencia entre los rangos de las variables correspondientes,
- $n$ es el número de observaciones.

```python

from scipy import stats

# create two variables
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([2, 3, 4, 5, 6, 7])

# calculate Spearman's rank correlation coefficient
rho, _ = stats.spearmanr(x, y)

print("Spearman rank correlation coefficient is", rho)
```

### 4. Coeficiente de correlación de rangos de Kendall

El coeficiente de correlación de rangos de Kendall (τ, también conocido como tau de Kendall) es otra medida no paramétrica de la fuerza y la dirección de la asociación que existe entre dos variables medidas en una escala ordinal o continua.

La τ de Kendall se calcula como:

$$
\tau = \frac{{n_c - n_d}}{{\sqrt{{(n(n-1))/2}}}}
$$

donde:
- $n_c$ es el número de pares concordantes,
- $n_d$ es el número de pares discordantes.


```python
from scipy import stats

# create two variables
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([2, 3, 4, 5, 6, 7])

# calculate Kendall's τ
tau, _ = stats.kendalltau(x, y)

print("Kendall rank correlation coefficient is", tau)
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt

# Load the 'Boston House Prices' dataset
boston = load_boston()
df = pd.DataFrame(boston.data, columns=boston.feature_names)
df['MEDV'] = boston.target

# Define the variables
x = df['LSTAT'].values
y = df['MEDV'].values
```

```python
# Scatter plot of the two variables
plt.figure(figsize=(12, 6))
plt.scatter(x, y, alpha=0.5)
plt.title('Scatter plot of LSTAT vs MEDV')
plt.xlabel('LSTAT')
plt.ylabel('MEDV')

# Linear fit (1st degree polynomial)
linear_fit = np.polyfit(x, y, 1)
linear_fit_line = np.poly1d(linear_fit)
plt.plot(x, linear_fit_line(x), label='linear fit', color='red')

# Polynomial fit (2nd degree polynomial)
polynomial_fit = np.polyfit(x, y, 2)
polynomial_fit_line = np.poly1d(polynomial_fit)
plt.plot(sorted(x), polynomial_fit_line(sorted(x)), label='polynomial fit', color='green')

plt.legend()
plt.show()
```

```python
# Calculate Pearson's correlation coefficient
N = len(x)
r = (N*np.sum(x*y) - np.sum(x)*np.sum(y)) / np.sqrt((N*np.sum(x**2) - np.sum(x)**2)*(N*np.sum(y**2) - np.sum(y)**2))
print("Pearson correlation coefficient is", r)

# Calculate Spearman's rank correlation coefficient
rho, _ = stats.spearmanr(x, y)
print("Spearman rank correlation coefficient is", rho)

# Calculate Kendall's tau
tau, _ = stats.kendalltau(x, y)
print("Kendall rank correlation coefficient is", tau)

```


###  Correlación frente a causalidad

Un concepto importante que hay que recordar al interpretar los coeficientes de correlación es que la correlación no implica causalidad. Es decir, el hecho de que dos variables estén correlacionadas no significa que los cambios en una de ellas provoquen cambios en la otra. Podría haber otros factores subyacentes que influyeran en ambas variables, o la relación podría ser casual.

### Correlación y regresión lineal

La regresión lineal y la correlación son dos conceptos estadísticos estrechamente relacionados. He aquí cómo:

#### Correlación

La correlación cuantifica el grado de relación entre dos variables. La correlación no ajusta una línea a través de los datos. Con la correlación, no tienes que preocuparte de qué variable es independiente y cuál es dependiente. Simplemente tienes dos variables y quieres saber si están relacionadas.

#### Regresión lineal

En cambio, la regresión lineal se utiliza para predecir el valor de una variable dependiente a partir del valor de una o varias variables independientes. Se calcula la línea de mejor ajuste y se obtiene una ecuación que puede utilizarse para predecir valores (dentro del intervalo de los datos). La pendiente y el intercepto de la recta de regresión se determinan minimizando la suma de los cuadrados de las distancias verticales de los puntos a la recta.

#### La conexión

El cuadrado del coeficiente de correlación de Pearson (r²) es la proporción de la varianza de una variable que puede predecirse a partir de la otra. Es lo mismo que el "coeficiente de determinación" que se ve al determinar en una regresión.

En una regresión lineal simple, en la que tenemos una única variable independiente y una variable dependiente, el coeficiente de correlación es igual a la raíz cuadrada del R-cuadrado, y comparten el mismo signo. Esto significa que si realizamos una regresión lineal y descubrimos que el valor de R-cuadrado es 0,49, la correlación (r) sería -0,7 o 0,7 dependiendo de la dirección de la relación.

Sin embargo, en el caso de la regresión múltiple, con más de una variable independiente, no existe una relación tan simple entre el coeficiente de correlación y la R-cuadrado múltiple. Cada variable independiente tendrá su propio coeficiente de correlación con la variable dependiente, pero juntas no forman directamente el R-cuadrado de toda la regresión.



```python
from sklearn.linear_model import LinearRegression

# Reshape the array
x = df['LSTAT'].values.reshape(-1,1)
y = df['MEDV'].values.reshape(-1,1)

# Perform the linear regression
model = LinearRegression()
model.fit(x, y)

# R-squared value
r_sq = model.score(x, y)
print('Coefficient of Determination (R-squared):', r_sq)
```

Este código ajusta un modelo lineal a los datos e imprime el valor R-cuadrado, que nos indica la proporción de la varianza en la variable dependiente que es predecible a partir de la(s) variable(s) independiente(s). En nuestro caso, nos dice cuánta de la variación en `MEDV` puede ser explicada por `LSTAT`.
